Workspace: Recopilando información del área de trabajoEste proyecto es una aplicación web construida con Next.js y TypeScript, diseñada para la gestión y firma digital de documentos. Utiliza Supabase como backend para la autenticación, base de datos y almacenamiento de archivos. El estilizado se realiza con Tailwind CSS y parece emplear componentes de UI de Shadcn/ui.

Aquí tienes un desglose de las partes principales del proyecto:

1.  **app**: Contiene la lógica principal de las páginas y rutas de la aplicación.
    *   **layout.tsx**: Define la estructura de diseño global, incluyendo proveedores como `ThemeProvider` para la gestión de temas y `SupabaseProvider` para la integración con Supabase.
    *   **page.tsx**: Es la página de inicio o landing page de la aplicación, implementada en landing-page.tsx.
    *   **`dashboard/`**: Agrupa las páginas accesibles tras el inicio de sesión, como:
        *   `page.tsx`: El panel principal del dashboard.
        *   `documents/`: Para listar (`app/dashboard/documents/page.tsx`) y ver detalles de documentos ([`app/dashboard/documents/[id]/page.tsx`](app/dashboard/documents/[id]/page.tsx)).
        *   `users/`: Para la gestión de usuarios (`app/dashboard/users/page.tsx`) y creación de nuevos usuarios (`app/dashboard/users/create/page.tsx`).
        *   `area-users/`: Para que los coordinadores de área gestionen usuarios de su departamento (`app/dashboard/area-users/page.tsx`) e inviten nuevos (`app/dashboard/area-users/invite/page.tsx`).
        *   `sign/`: Página para listar documentos pendientes de firma (`app/dashboard/sign/page.tsx`).
        *   `signed/`: Página para ver documentos ya firmados por el usuario (`app/dashboard/signed/page.tsx`).
        *   `profile/`: Para que el usuario gestione su perfil (`app/dashboard/profile/page.tsx`).
    *   **page.tsx**: Página de inicio de sesión.
    *   **page.tsx**: Página de registro de nuevos usuarios (actualmente visible).
    *   **page.tsx**: Parece ser una página pública o alternativa para visualizar documentos firmados.

2.  **components**: Alberga los componentes reutilizables de la interfaz de usuario.
    *   **`auth/`**: Componentes para autenticación, como `LoginForm` y `RegisterForm`.
    *   **`documents/`**: Componentes para la manipulación de documentos, como `PDFViewer` para visualizar PDFs, `DocumentSigningForm` para el proceso de firma, y `DocumentUploadButton` para subir nuevos documentos. También incluye `DocumentPermissionsManager` para gestionar quién puede acceder o firmar un documento.
    *   **`dashboard/`**: Componentes específicos del panel de control, como `DashboardHeader` (cabecera) y `DashboardNav` (menú de navegación).
    *   **`users/`**: Componentes para la gestión de usuarios, como `CreateUserForm` y `InviteUserForm`.
    *   **`profile/`**: Componentes para la edición del perfil de usuario, como `ProfileForm`.
    *   **`ui/`**: Componentes genéricos de UI (botones, tarjetas, modales, etc.), como `Button`, `Card`, `Dialog`, `Form`, `Input`, `Select`, `Toast`, `Sidebar`.
    *   **`landing-page.tsx`**: Componente que define el contenido de la página de inicio.
    *   **`theme-provider.tsx`**: Gestiona los temas de la aplicación (claro/oscuro).

3.  **hooks**: Contiene hooks personalizados de React.
    *   **`use-toast.ts`**: Para mostrar notificaciones (toasts).

4.  **lib**: Funciones de utilidad, configuración y lógica de negocio.
    *   **`supabase-browser.ts`** y `supabase-server.ts`: Clientes de Supabase para el navegador y el servidor, respectivamente.
    *   **`supabase-provider.tsx`**: Proveedor de contexto para Supabase.
    *   **`database.types.ts`**: Define los tipos de TypeScript para la estructura de la base de datos en Supabase.
    *   **`rbac.ts`**: Lógica para el Control de Acceso Basado en Roles (RBAC), definiendo roles (como `system_admin`, `area_coordinator`) y sus permisos.

5.  **public**: Archivos estáticos como imágenes y logos.

**Funcionalidades Clave:**

*   **Autenticación**: Registro e inicio de sesión de usuarios utilizando Supabase.
*   **Gestión de Documentos**: Subida, visualización y descarga de documentos (principalmente PDF).
*   **Firma Digital**: Proceso para que los usuarios firmen digitalmente los documentos.
*   **Gestión de Usuarios**: Creación e invitación de usuarios, con diferentes roles y permisos.
*   **Control de Acceso Basado en Roles (RBAC)**: El sistema restringe el acceso a ciertas funcionalidades según el rol del usuario.
*   **Gestión de Perfil**: Los usuarios pueden actualizar su información personal.

En resumen, es un sistema robusto para el ciclo de vida de documentos que requieren firmas digitales, con un panel de administración para gestionar usuarios y documentos, y con diferentes niveles de acceso según el rol del usuario.
